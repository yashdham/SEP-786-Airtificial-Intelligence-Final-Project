```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.neighbors import KNeighborsClassifier
import pandas as pd
from sklearn.decomposition import PCA
```

```python
data = pd.read_csv(r'C:\Users\Yash\Desktop\Fall 2019 Material\AI\mobiledatatrain.csv')
data.info
x= data.iloc[:, :-1].values
y= data.iloc[:, -1].values
```

```python
#training and test data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.25, random_state =0)
```

```python
#feature scaling
from sklearn.preprocessing import StandardScaler 
sc_x = StandardScaler()
x_train = sc_x.fit_transform(x_train)
x_test = sc_x.transform(x_test)
```

```python
#SVR fitting
from sklearn.svm import SVR
reg = SVR(kernel = 'rbf')
reg.fit(x_train, y_train)

acc_svr = reg.score(x_test, y_test)

y_pred_svr = reg.predict(x_test)

er_svr = np.mean((y_test - y_pred_svr)**2)

er_svr

print ('Error ', er_svr)
print ('Accuracy ', acc_svr)
```

```python
#DecisionTreeREgressor fitting
from sklearn.tree import DecisionTreeRegressor
regressor_tr = DecisionTreeRegressor()
regressor_tr.fit(x_train, y_train)

acc = regressor_tr.score(x_test, y_test)

y_pred_tr = regressor_tr.predict(x_test)

er_tr = np.mean((y_test - y_pred_tr)**2)

print ('Error ', er_tr)
print ('Accuracy ', acc)
```

```python
#RandomForestRegrossor fitting
from sklearn.ensemble import RandomForestRegressor
regressor_rf = RandomForestRegressor(n_estimators = 10)
regressor_rf.fit(x_train, y_train)

acc_rf = regressor_rf.score(x_test, y_test)

y_pred_rf = regressor_rf.predict(x_test)

er_rf = np.mean((y_test - y_pred_rf)**2)

er_rf

print ('Error ', er_rf)
print ('Accuracy ', acc_rf)
```
